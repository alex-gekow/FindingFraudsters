```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng(216)
## randomly generated data
x = rng.uniform(-6,6, 500) 
y = 3*x+ x**2 + 5*rng.standard_normal(500)
```

```python
## Visualize that data
plt.figure(figsize=(8,6))

plt.scatter(x, y)

plt.xlabel("$x$", fontsize=12)
plt.ylabel("$y$", fontsize=12)

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
slr = LinearRegression(copy_X=True)

slr.fit(x.reshape(-1,1), y)

residuals = y - slr.predict(x.reshape(-1,1))
```

```python
## Visualize that data
plt.figure(figsize=(8,6))

plt.scatter(x, y, 
            label="Observed Data",
            alpha=.6)

plt.plot(np.linspace(-5,6,10),
         slr.predict(np.linspace(-5,6,10).reshape(-1,1)),
         "k",
         label="SLR Line")

plt.xlabel("$x$", fontsize=12)
plt.ylabel("$y$", fontsize=12)

plt.legend(fontsize=12)

plt.show()
```

```python
## Make a residual plot
plt.figure(figsize=(8,6))

plt.scatter(x, residuals)

plt.xlabel("$x$", fontsize=12)
plt.ylabel("residuals", fontsize=12)


plt.show()
```

```python
## fit the correct model
reg = LinearRegression(copy_X=True)

X = np.zeros((len(x),2))
X[:,0] = x
X[:,1] = x**2

reg.fit(X,y)

residuals = y - reg.predict(X)

## Visualize that data
plt.figure(figsize=(8,6))

plt.scatter(x, y, 
            label="Observed Data",
            alpha=.6)


inp = np.zeros((10,2))
inp[:,0] = np.linspace(-5,6,10).reshape(-1)
inp[:,1] = inp[:,0]**2

plt.plot(np.linspace(-5,6,10),
         reg.predict(inp),
         "k",
         label="SLR Line")

plt.xlabel("$x$", fontsize=12)
plt.ylabel("$y$", fontsize=12)

plt.legend(fontsize=12)

plt.show()
```
